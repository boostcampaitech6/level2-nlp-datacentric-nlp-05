```python
import os
import pandas as pd

BASE_DIR = os.getcwd()
DATA_DIR = os.path.join(BASE_DIR, '../data')
data = pd.read_csv(os.path.join(DATA_DIR, 'train.csv'))
data.head()
```

```python
from koeda import AEDA

aeda = AEDA(
    morpheme_analyzer="Okt", punc_ratio=0.3, punctuations=[".", ",", "!", "?", ";", ":"]
)
```

```python
text = 'SK텔레콤 해킹불가능 장거리 양자암호통신 성공'
aeda(text)
```

```python
aeda(text)
```

```python
def use_aeda(text):
    return aeda(text)
```

```python
aeda_df = data.copy()

aeda_df['text'] = aeda_df['text'].apply(use_aeda)
```

```python
aeda_df.head(10)
```

```python
aeda_train = pd.concat([data, aeda_df])
aeda_train
```

```python
aeda_train.reset_index(drop=True)
aeda_train.to_csv('../data/aeda_train.csv', index=False)
```
